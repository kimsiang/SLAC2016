```python
import ROOT
c = ROOT.TCanvas("myCanvasName","The Canvas Title",800,600)
```

```python
run=1932
myFile = ROOT.TFile('run{0}_beamTimeDiff2425.root'.format(run))
hist = myFile.Get("beamTimeDiff2425");
hist.SetTitle('run{}'.format(run))
graph1 = myFile.Get("Graph;1")
graph2 = myFile.Get("Graph;2")
#hist.Draw()
graph1.Draw("AP")
graph1.SetTitle('dT versus Energy, run{}'.format(run))
graph1.SetMarkerStyle(4)
graph1.SetMarkerColor(2)
graph1.GetXaxis().SetLimits(0,2500)
graph1.GetYaxis().SetRangeUser(-1.5,1.5)
graph1.GetXaxis().SetTitle("Energy of xtal24 or xtal25 [#photon]")
graph1.GetYaxis().SetTitle("dT=T_{xtal24}-T_{xtal25}")
graph2.Draw("Psame")
graph2.SetMarkerStyle(4)
graph2.SetMarkerColor(4)
ROOT.enableJSVis()


legend = ROOT.TLegend(0.4,0.75,0.6,0.85)
legend.AddEntry(graph1, "xtal24","p")
legend.AddEntry(graph2, "xtal25","p")
legend.SetBorderSize(0)
legend.Draw()
c.Draw()
c.Print('run{}.png'.format(run))
```
